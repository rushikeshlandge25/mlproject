```python
import pandas as pd
df=pd.read_csv("D:\Rushikesh Landge\datasets\diamonds.csv")
df=df.drop('id', axis=1)
X = df.drop('price', axis=1)
y = df['price']
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

```

```python

from sklearn.linear_model import LinearRegression
numerical_column = X.select_dtypes(exclude=object).columns
categorical_column = X.select_dtypes(include='object').columns

# Correct ordinal order (worst → best)
cut_cat = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_cat = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
clarity_cat = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score

num_pipe = Pipeline([
    ('imputer', SimpleImputer()),
    ('scaling', StandardScaler())
])

cat_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoding', OrdinalEncoder(
        categories=[cut_cat, color_cat, clarity_cat],handle_unknown='error'

    ))
])

col_transform = ColumnTransformer([
    ('num_pipeline', num_pipe, numerical_column),
    ('cat_pipeline', cat_pipe, categorical_column),
    
])

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
lr =LinearRegression()
rf=RandomForestRegressor(
    n_estimators=100,
    max_depth=5,
    n_jobs=-1,
    random_state=42,

)

Pipe3=Pipeline([
    ('preprosessdata',col_transform),
    ('model',rf)
])
```

```python
Pipe3.fit(x_train,y_train)
```

```python
x_train
```

```python
y_pred=Pipe3.predict(x_test)
print(r2_score(y_test ,y_pred))
```
